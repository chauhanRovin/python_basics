2. Check String Prefix/Suffix

```python
filename = "sales_data.csv"

print(filename.endswith(".csv"))
print(filename.startswith("sales"))
```

3. Find Text

```python
sentence = "Python is easy to learn"

print(sentence.find("easy"))
print(sentence.find("hard"))
```

4. Count Occurrences

```python
sentence = "data data analytics data"

print(sentence.count("data"))
```

5. Check String Type

```python
print("123".isdigit())
print("Python".isalpha())
print("Python123".isalnum())
```
