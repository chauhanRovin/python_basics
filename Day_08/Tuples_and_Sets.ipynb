# Tuples

A tuple is an ordered and immutable collection.

Create a Tuple

```python
fruits = ("apple", "mango", "banana")

print(fruits)
print(type(fruits))
```

Access Elements

```python
print(fruits[0])
print(fruits[2])
```

Count and Index

```python
print(fruits.count("mango"))
print(fruits.index("mango"))
```

# Sets

A set stores unique values

```python
numbers = {10, 20, 30, 40}

print(numbers)

numbers.add(50)
print(numbers)

numbers.remove(20)
print(numbers)
```

```python
marks = [80, 90, 80, 75, 90, 95]

unique_marks = set(marks)

print(unique_marks)
```

Practice Questions

```python
cities = ("Delhi", "Mumbai", "Pune", "Jaipur", "Lucknow")

print(cities[0])
print(cities[-1])
```

```python
numbers = {1, 2, 2, 3, 4, 4, 5}

unique_num = set(numbers)

print(unique_num)
```

```python
data = {10, 20, 20, 30, 30, 30, 40}

print(len(set(data)))
```
