#Loops in Python
  Loops are used to execute a block of code repeatedly until a condition is met or all items in a sequence are processed.
  The main types are :
  a)For loops (iterating over sequences)
  b)While loops (executing code based on a condition)

Examples of for loop

```python
for i in range(5):
  print(i)
```

```python
for i in range(1,6):
  print(i)
```

Examples of while loop

```python
count = 0

while(count <=5 ):
  print(count)
  count += 1
```

###Practice Questions
Question 1: Print Numbers 1 to 10

```python
# Using for loop
for i in range(1, 11):
  print(i)

# Using while loop
count = 1
while (count <= 10):
  print(count)
  count += 1
```

Question 2: Print Even Numbers from 1 to 20

```python
for i in range(2, 20, 2):
  print(i)
```

Question 3: Multiplication Table of 5

```python
for i in range(1, 11):
  print(5*i)
```

Question 4:Sum of First 10 Numbers

```python
sum = 0

for i in range(11):
  sum = sum+i

print("Required_Sum = ",sum)
```

Question 6:Count vowels in a word

```python
word = "analytics"
count = 0

for char in word:
  if char in 'aeiou':
    count += 1

print("Vowels : ", count)
```
