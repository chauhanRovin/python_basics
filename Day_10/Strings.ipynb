1. Creating Strings

```python
name = "Rovin Chauhan"

print(name)
print(type(name))
```

2. Indexing

```python
text = "Python"

print(text[0])
print(text[2])
print(text[5])
```

3. Slicing

```python
text = "DataAnalytics"

print(text[0:4])
print(text[0:4:2])
print(text[:4])
print(text[0::2])
```

4. Useful String Methods

```python
text = "python for data analytics"

print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
```

5. Replace

```python
sentence = "I love Java"
print(sentence.replace("Java", "Python"))
```

6. Split

```python
skills = "Python,SQL,Power BI"

print(skills.split(","))
```

7. Join

```python
words = ["Data", "Analyst"]

print(" ".join(words))
```
